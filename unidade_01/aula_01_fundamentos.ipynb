### Seção transversal da imagem

```python
l500 = mandruva[500,:]

plt.plot(np.arange(1024), l500, '-k')
```

### Mudança da quantização

```python
# Quantização em 2 níveis de cinza (BINARIZAÇÃO)

mandruva_02 = mandruva.copy()
mandruva_02[mandruva_02 > 128] = 255
mandruva_02[mandruva_02 < 255] = 0

plt.imshow(mandruva_02, cmap='gray', vmin=0, vmax=255)
```

### Mudanças na amostragem

```python
mandruva_03 = mandruva.copy()
n = 16 # parametro da subamostragem

for i in range (0, 1024, n):
    for j in range (0, 1024, n):
        mandruva_03[i:i+n,j:j+n] = mandruva[i,j]

plt.imshow(mandruva_03, cmap='gray', vmin=0, vmax=255)
```

```python
l500 = mandruva_03[500,:]

plt.plot(np.arange(1024), l500, '-k')
```
