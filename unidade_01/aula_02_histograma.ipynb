# O histograma de nível de cinza

## Distribuição de frequência dos valores de nível de cinza

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from scipy.fft import dct
from skimage import color
from skimage import io
```

```python
zero = io.imread('zero.jpg')

# construção do histograma da imagem do zero

h = np.histogram(zero.ravel(), bins=256, range=(0, 255))
```

```python
# visualização da imagem e de seu histograma

plt.imshow(zero, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.plot(h[0], '-k')
plt.show()
```

```python
mini_mapa = io.imread('mini_mapa.jpg')

# construção do histograma do mini mapa

h = np.histogram(mini_mapa.ravel(), bins=256, range=(0, 255))

# visualização da imagem e de seu histograma

plt.imshow(mini_mapa, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.plot(h[0], '-k')
plt.show()
```

## Operações com o histograma

```python
# negativo

zero_negativo = 255 - zero

h = np.histogram(zero_negativo.ravel(), bins=256, range=(0, 255))

# visualização da imagem e de seu histograma

plt.imshow(zero_negativo, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.plot(h[0], '-k')
plt.show()
```

```python
# raiz quadrada

zero_raiz = np.sqrt(zero)
zero_raiz = 255 * zero_raiz / zero_raiz.max()
zero_raiz = zero_raiz.astype('uint8')

# visualização da imagem e de seu histograma

plt.imshow(zero_raiz, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.plot(h[0], '-k')
plt.show()
```

```python
carro = io.imread('carro.jpg')

plt.imshow(carro, cmap='gray', vmin=0, vmax=255)
```

```python
h = np.histogram(carro.ravel(), bins=256, range=(0,255))

plt.plot(h[0], '-k')
plt.show()
```

```python
carro_raiz = np.sqrt(carro)
carro_raiz = carro_raiz - carro_raiz.min()
carro_raiz = (255 * carro_raiz / carro_raiz.max()).astype('uint8')

plt.imshow(carro_raiz, cmap='gray')
```

```python
h = np.histogram(carro_raiz.ravel(), bins=256, range=(0,255))

plt.plot(h[0], '-k')
plt.show()
```

```python
matheus = io.imread('matheus.jpg')

print(matheus.dtype, matheus.shape, matheus.max())

plt.imshow(matheus)
```

```python
img = (color.rgb2gray(matheus)*255).astype('uint8')

print(img.dtype, img.shape, img.max())

plt.imshow(img, cmap='gray')
```

```python
# calculo da dct da imagem do Matheus

dct_matheus = np.abs(dct(dct(img.T).T))

# construcao do histograma da imagem

h = np.histogram(dct_matheus.ravel(), bins=256)

# visualizacao da imagem e do histograma

dct_matheus = dct_matheus - dct_matheus.min()
dct_matheus = dct_matheus * 255 / dct_matheus.max()
dct_matheus = dct_matheus.astype('uint8')

plt.imshow(dct_matheus, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.plot(h[0], '-k')
plt.show()
```

```python
# aplicacao de log
# calculo da dct da imagem

dct_matheus = dct(dct(img.T).T)

log_matheus = np.log(1 + (np.abs(dct_matheus)))
log_matheus = log_matheus - log_matheus.min()
log_matheus = log_matheus * 255 / log_matheus.max()
log_matheus = log_matheus.astype('uint8')

plt.imshow(log_matheus, cmap='gray')
plt.show()
```
